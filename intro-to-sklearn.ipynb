## Six steps to train a model using scikitlearn
1. Getting the data ready
2. Choose the right estimator/algorithm for the problem
3. Fit the model/algorithm and use it to make predictions on the data
4. Evaluating a model
5. Improve a model
6. Save and load a trained model

```python
# 1. Get the data ready
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# use the columns to predict the target, heart disease or not, 1 or 0.

# First thing is to get the X which is the feature matrix, i.e the column,
# headings and y kinda the last column

# Create x (features matrix or feature variables or data)
x = heart_disease.drop("target", axis = 1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right model and hyperparameters.
# To predict whether a patient has a heart disease or not,
# we use a classification algorithm.
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
# we'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split
# split the data to train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# 80% of the data will be used for training because test_size 0.2 = 20%

clf.fit(x_train, y_train);
x_train
```

```python
# Make a prediction (still step 3)
x_test
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
# other metrics to use check eval. score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators = i).fit(x_train, y_train)
    print(f"Model accuracy on test set:{clf.score(x_test, y_test)*100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it
# We can save a model with python pickel
import pickle
pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
# Import our model/file back
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(x_test, y_test)
```
